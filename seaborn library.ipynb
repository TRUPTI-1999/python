### bar plot

### distplot

creates the distribution of any continuos data lets create the distribution of miles per model of the different vehicles of the cars the mpg metrics measures the total distance the car can travel per model of the car

```python
sns.distplot(df.mpg, bins=10,color="g")
```

### heatmap()
