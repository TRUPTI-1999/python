### Conditional Charges

```python
details={"Name":["Ankit","Aishwarya","Ankit","Aishwarya","Shaurya","Shivangi"],
         "math":[80,40,50,70,42,61],
         "sci":[83,21,43,61,72,71],
         "Hindi":[53,21,73,11,92,11],
         "chem":[73,41,43,41,82,91],
         "Bio":[93,41,23,41,45,51],
         "sanskrit":[73,81,53,41,72,21]}
```

```python
details
```

```python
d1=pd.DataFrame(details)
```

```python
d1
```

```python
d1["total"]=0
```

```python
d1
```

```python
d1["total"]=d1["math"]+d1["sci"]+d1["Hindi"]+d1["chem"]+d1["Bio"]+d1["sanskrit"]
d1
```

```python
d1["grade"]=0
```

```python
d1
```

```python
d1["%"]=(d1["total"]/600)*100
d1
```

- less than 45-Fail
- 45-60 pass
- 60-70 first class
- 70-distincttion

```python
d1.loc[d1["%"]<45]
d1.loc[d1["%"]<45,"grade"]="Fail"
```

```python
d1
```

```python
d1.loc[(d1["%"]>45)&(d1["%"]<60)]
d1.loc[(d1["%"]>45)&(d1["%"]<60),"grade"]="pass"
```

```python
(d1["%"]>45)&(d1["%"]<60)
```

```python
d1
```

```python
d1.loc[(d1["%"]>70)]
d1.loc[(d1["%"]>70),"grade"]="Distination"
d1.loc[(d1["%"]>60) & (d1["%"]<70)]
d1.loc[(d1["%"]>60)& (d1["%"]<70),"grade"]="First class"
```
