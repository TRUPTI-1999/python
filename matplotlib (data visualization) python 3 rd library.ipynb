### Matplotlib Markers


Matplotlib Markers
- In this section we see more about Markers,
- We can use the keyword argument marker to emphasize each point with a specified marker
- Eg:
    Mark each point with a circle.

```python
ypoints=np.array([3,8,1,10])
plt.plot(ypoints,marker="o")
plt.show()
```

```python
ypoints=np.array([3,8,1,10])
plt.plot(ypoints,marker='*')
plt.show()
```

## Format String fmt

- You can use also use the shortcut string notation parameter to specify the marker.
- This parameter is also called ft, and is written with this syntax:
    Marker|line color.
- Eg: Mark each point with a circle.

```python
ypoints=np.array([3,8,1,10])
plt.plot(ypoints,'*:y')
plt.show
```

```python
#This will be like that string range giving, so like that we can simply tell the python 
#do give a plot with line as this, marker as this like that. 
```

- The marker value can be anything from the marker Reference above.
- The line value can be one of the following

```python
Line Syntax                 Description
'_'                           Solid  line
':'                            Dotted Line
'_ _'                          Dashed Line
'-.'                           Dashed/dotted line
```

```python
# if you leave out the line value in the fmt parameters,no line will be plotted.
```

## Marker Size

- you can use the keyword argument markersize or the shorter varsion,ms to set the size of the marker:

- eg
  set the size of the markers to 20

```python
ypoints=np.array([3,8,1,10])
plt.plot(ypoints,marker='o',ms=20)
plt.show()
```

## Matplotlib Labels and Title

- with pyplot, we can use the xlabel() and ylabel() function to set a label for the x- and y-axis.

```python
x=np.array([80,85,90,95,100,105,110,115,120,125])
y=np.array([240,250,260,270,280,290,300,310,320,330])
plt.plot(x,y)
plt.xlabel("Average pulse")
plt.ylabel("calorie Burnage")
plt.show()
```

## Create a Title for a plot

- with pyplot, you can use the title() function to set a title for the plot.

```python
x=np.array([80,85,90,95,100,105,110,115,120,125])
y=np.array([240,250,260,270,280,290,300,310,320,330])
plt.plot(x,y)
plt.title("sport watch Data")
plt.xlabel("Average pulse")
plt.ylabel("calorie Burnage")
plt.show()
```

### Set Font Properties for Title and labels

- we can use the fontdict parameter in xlabel(),ylabel(),and title() to set font properties for the title and labels

```python
x=np.array([80,85,90,95,100,105,110,115,120,125])
y=np.array([240,250,260,270,280,290,300,310,320,330])
font1={'family':'serif','color':'blue','size':20}
font2={'family':'serif','color':'darkred','size':15}
plt.title("sports Watch data",fontdict=font1)
plt.xlabel("Average Burnage",fontdict=font2)
plt.ylabel("Calorie burnage",fontdict=font2)

plt.plot(x,y)
plt.show()
```

## Position the Title

- we can use the loc parameter in the ()to position the title.
- Lagel values are:'left','right', and 'center'.Default value is 'center'.

```python
x=np.array([80,85,90,95,100,105,110,115,120,125])
y=np.array([240,250,260,270,280,290,300,310,320,330])
plt.title("sport watch data",loc='left')
plt.xlabel("Average pulse")
plt.ylabel("Calorie Burnage")
plt.plot(x,y)
plt.show()
```

### Matplotlib Adding Grid Lines

With pyplot,you can use the grid() function to add grid lines to the plot.

```python
x=np.array([80,85,90,95,100,105,110,115,120,125])
y=np.array([240,250,260,270,280,290,300,310,320,330])
plt.title("sports watch Data")
plt.xlabel("Average pulse")
plt.ylabel("calorie Burnage")
plt.plot(x,y)
plt.grid()
plt.show()
```

### The subplots () Function

The subplots() Function

The subplots() function takes three arguments that describes the layout of the figure.

The layout is organized in rows and columns, which are represented by the first and second argument.

The third argument represents the index of the current plot.

pit. subplot(1, 2, 1)
#the figure has 1 row, 2 columns, and this plot is the first plot.

plt. subplot(1, 2, 2)
#the figure has 1 row, 2 columns, and this plot is the second plot.

So, if we want a figure with 2 rows an 1 column (meaning that the two plots will be displayed on top of each other instead of side-by-side), we can write the syntax like this:

```python
x=np.array([0,1,2,3])
y=np.array([3,8,1,10])
plt.subplot(2,3,1)
plt.subplot(2,3,1)
plt.plot(x,y)
x=np.array([0,1,2,3])
y=np.array([10,20,30,40])
plt.subplot(2,3,2)
plt.plot(x,y)
x=np.array([0,1,2,3])
y=np.array([3,8,1,10])
plt.subplot(2,3,3)
plt.plot(x,y)
```
